```python
import cv2
import matplotlib.pyplot as plt
```

```python
gambar = cv2.imread("/content/IMG_20210704_110719_575.jpg")
```

```python
plt.imshow(gambar)
```

```python
def toRGB(imgs):
  imgs = cv2.cvtColor(imgs, cv2.COLOR_BGR2RGB)
  return imgs
```

```python
def rgb_to_gray(img):
  height, width, channels = img.shape

  for i in range(height):
    for j in range(width):
      img[i,j] = 0.3 * img[i,j][0]+ 0.59 * img[i,j][1]+ 0.11*img[i,j][2]

      return img
```

```python
gambarWarna = toRGB(gambar)
gambarAbu = rgb_to_gray(gambar)
```

```python
plt.imshow(gambarWarna)
```

```python
plt.imshow(gambarAbu)
```

```python
numDownSampling = 2
numBilateralFilter = 7
```

```python
for _ in range(numDownSampling):
  gambarWarna = cv2.pyrDown(gambarWarna)
  gambar1 = gambarWarna
```

```python
plt.imshow(gambar1)
```

```python
 for _ in range(numBilateralFilter):
   gambarWarna = cv2.bilateralFilter(gambar1, d=9, sigmaColor = 9, sigmaSpace = 7)
   gambar2 = gambarWarna
```

```python
plt.imshow(gambar2)
```

```python
for _ in range(numDownSampling):
  gambarWarna = cv2.pyrUp(gambar2)
  gambar3 = gambarWarna
```

```python
plt.imshow(gambar3)
```

```python
gambar4 = gambarAbu
gambar_blur = cv2.medianBlur(gambarAbu, 5).astype('uint8')
gambar5 = gambar_blur
```

```python
plt.imshow(gambar4)
```

```python
plt.imshow(gambar5)
```

```python
laplacian_image = cv2.Laplacian(gambar5, ddepth=-1, ksize = 5)
ret,thresh = cv2.threshold(laplacian_image,125,255,cv2.THRESH_BINARY_INV)
gambar6 = thresh
```

```python
plt.imshow(laplacian_image)
```

```python
plt.imshow(gambar6)
```

```python
gambarWarnaBaru = gambar3
gambarWarnaBaru = cv2.resize(gambarWarnaBaru, (1025,680))
```

```python
gambar6 = cv2.resize(gambar6, (1025, 680))
```

```python
cartoon = cv2.bitwise_and(gambarWarnaBaru, gambar6)
plt.imshow(cartoon)
```
